```sql
-- Month-over-Month Sales Performance Analysis
-- Calculate percentage change in sales between current and previous months

SELECT *,
  currentmonthSales - previousmonthSales AS MOM_Change,
  ROUND(CAST((currentmonthSales - previousmonthSales) AS FLOAT)/previousmonthSales*100, 1) AS MOM_Perc
FROM 
(
  SELECT 
    MONTH(orderdate) AS ordermonth, 
    SUM(sales) AS currentmonthSales, 
    LAG(SUM(sales)) OVER(ORDER BY MONTH(orderdate)) AS previousmonthSales 
  FROM salesdb.orders 
  GROUP BY MONTH(orderdate)
) t;
```

```sql
-- Customer Loyalty Analysis
-- Rank customers based on average days between orders (lower = more loyal)

SELECT 
  customerid,
  AVG(Daysuntilnxtorder) AS avgdays,
  RANK() OVER(ORDER BY COALESCE(AVG(Daysuntilnxtorder), 99999)) AS RankAvg
FROM (
  SELECT 
    OrderID, 
    customerid, 
    orderdate AS currentorder,
    LEAD(orderdate) OVER(PARTITION BY customerid ORDER BY orderdate) AS NextOrder,
    DATEDIFF(day, orderdate, LEAD(orderdate) OVER(PARTITION BY customerid ORDER BY orderdate)) AS Daysuntilnxtorder
  FROM salesdb.orders
) t
GROUP BY customerid
ORDER BY RankAvg;
```

```sql
-- Comprehensive Customer Analysis using Multiple CTEs
-- Combines total sales, last order date, customer ranking, and segmentation

WITH cte_totalsales AS (
  SELECT 
    customerid,
    SUM(sales) AS totalsales
  FROM salesdb.orders
  GROUP BY customerid
),
CTE_Last_Order AS (
  SELECT 
    customerid,
    MAX(orderdate) AS lastorder
  FROM salesdb.orders
  GROUP BY customerid
),
CTE_customer_rank AS (
  SELECT 
    customerid,
    totalsales,
    RANK() OVER(ORDER BY totalsales DESC) AS customerrank
  FROM cte_totalsales
),
cte_customer_segments AS (
  SELECT 
    customerid, 
    totalsales,
    CASE
      WHEN totalsales > 100 THEN 'High'
      WHEN totalsales > 80 THEN 'Medium'
      ELSE 'Low'
    END AS customersegment
  FROM cte_totalsales
)
SELECT 
  c.customerid,
  c.firstname,
  c.lastname,
  c.country,
  cts.totalsales,
  clo.lastorder,
  ccr.customerrank,
  ccs.customersegment
FROM salesdb.customers AS c
LEFT JOIN cte_totalsales AS cts ON cts.customerid = c.customerid
LEFT JOIN CTE_Last_Order AS clo ON clo.customerid = c.customerid
LEFT JOIN CTE_customer_rank AS ccr ON ccr.customerid = c.customerid
LEFT JOIN cte_customer_segments AS ccs ON ccs.customerid = c.customerid
ORDER BY cts.totalsales DESC;
```

```sql
-- Subquery in FROM clause: Find products with price higher than average

SELECT * 
FROM (
  SELECT 
    productid,
    product,
    price,
    AVG(price) OVER() AS avgprice
  FROM salesdb.products
) t 
WHERE price > avgprice;
```

```sql
-- Subquery in FROM clause: Rank customers based on total sales

SELECT *,
  RANK() OVER(ORDER BY totalsales DESC) AS CustomerRank
FROM (
  SELECT 
    customerid,
    SUM(sales) AS totalsales
  FROM salesdb.orders
  GROUP BY customerid
) t;
```

```sql
-- Subquery in SELECT clause: Show products with total order count

SELECT 
  productid,
  product,
  price,
  (SELECT COUNT(*) FROM salesdb.orders) AS TotalOrders  
FROM salesdb.products;
```

```sql
-- Subquery in JOIN clause: Show all customer details with total orders

SELECT c.*, o.totalorders
FROM salesdb.customers AS c
LEFT JOIN (
  SELECT 
    customerid,
    COUNT(*) AS totalorders
  FROM salesdb.orders
  GROUP BY customerid
) o ON c.customerid = o.customerid;
```

```sql
-- Subquery in WHERE clause: Find products with price higher than average

SELECT 
  productid,
  product,
  price 
FROM salesdb.products
WHERE price > (SELECT AVG(price) FROM salesdb.products);
```

```sql
-- Subquery with IN operator: Show orders made by customers in Germany

SELECT *
FROM salesdb.orders
WHERE customerid IN (
  SELECT customerid
  FROM salesdb.customers 
  WHERE country = 'Germany'
);
```

```sql
-- ANY operator equivalent: Female employees with salary greater than any male employee
-- (Salary > ANY(male salaries) means salary > minimum male salary)

SELECT 
  employeeid,
  firstname,
  lastname,
  gender,
  salary
FROM salesdb.employees
WHERE gender = 'F'
  AND salary > (
    SELECT MIN(salary) 
    FROM salesdb.employees 
    WHERE gender = 'M'
  );
```

```sql
-- ALL operator equivalent: Female employees with salary greater than all male employees
-- (Salary > ALL(male salaries) means salary > maximum male salary)

SELECT 
  employeeid,
  firstname,
  lastname,
  gender,
  salary
FROM salesdb.employees
WHERE gender = 'F'
  AND salary > (
    SELECT MAX(salary) 
    FROM salesdb.employees 
    WHERE gender = 'M'
  );
```

```sql
-- Single CTE: Find total sales per customer with customer details

WITH cte_totalsales AS (
  SELECT 
    customerid,
    SUM(sales) AS totalsales
  FROM salesdb.orders
  GROUP BY customerid
)
SELECT 
  c.customerid,
  c.firstname,
  c.lastname,
  c.country,
  cts.totalsales
FROM salesdb.customers AS c
LEFT JOIN cte_totalsales AS cts ON cts.customerid = c.customerid;
```

```sql
-- Multiple CTEs: Find total sales and last order date per customer

WITH cte_totalsales AS (
  SELECT 
    customerid,
    SUM(sales) AS totalsales
  FROM salesdb.orders
  GROUP BY customerid
),
CTE_Last_Order AS (
  SELECT 
    customerid,
    MAX(orderdate) AS lastorder
  FROM salesdb.orders
  GROUP BY customerid
)
SELECT 
  c.customerid,
  c.firstname,
  c.lastname,
  cts.totalsales,
  clo.lastorder
FROM salesdb.customers AS c
LEFT JOIN cte_totalsales AS cts ON cts.customerid = c.customerid
LEFT JOIN CTE_Last_Order AS clo ON clo.customerid = c.customerid;
```
